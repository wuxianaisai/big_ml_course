**Сводные таблицы в Pandas**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Data/Sales_Funnel_CRM.csv")
```

```python
df
```

```python
licenses = df[["Company", "Product", "Licenses"]]
```

```python
licenses
```

```python
pd.pivot(data=licenses, index="Company", columns="Product", values="Licenses")
```

```python
pd.pivot_table(df, index="Company", aggfunc=sum).select_dtypes(include="number").drop("Account Number", axis=1)
```

```python
df.groupby("Company").sum().select_dtypes(include="number")
```

```python
pd.pivot_table(df, index=["Account Manager", "Contact"], columns=["Product"], values=["Sale Price"], aggfunc=sum, fill_value=0)
```

```python
pd.pivot_table(df,
               index=["Account Manager", "Contact"],
               columns=["Product"],
               values=["Sale Price"],
               aggfunc=[np.sum, np.mean],
               fill_value=0,
               margins=True)
```
