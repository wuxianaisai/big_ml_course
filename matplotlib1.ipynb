**Основы Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
plt.plot(x, y);
plt.xlabel("Ось x")
plt.ylabel("Ось y")
plt.xlim(0, 6)
plt.ylim(0, 15)
plt.title("Название графика")
plt.savefig("MyFirstPlot.png")
```
