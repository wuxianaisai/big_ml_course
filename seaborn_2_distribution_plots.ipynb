**Distribution Plots Ð² Seaborn**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Data/dm_office_sales.csv")
```

```python
df.head()
```

```python
sns.rugplot(x="salary", data=df)
```

```python
sns.set(style="darkgrid")
sns.displot(x="salary", data=df, bins=20, color="red", edgecolor="blue", linewidth=4, kde=True)
```

```python
sns.histplot(x="salary", data=df)
```

```python
sns.kdeplot(x="salary", data=df)
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0, 100, 200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=["age"])
```

```python
sns.rugplot(data=sample_ages, x="age")
```

```python
sns.displot(data=sample_ages, x="age", rug=True, bins=30, kde=True)
```

```python
sns.kdeplot(data=sample_ages, x="age", clip=[0, 100], fill=True)
```
