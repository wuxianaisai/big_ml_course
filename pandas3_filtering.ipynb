**Выборка данных по условию (Conditional Filtering) в Pandas**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Data/tips.csv")
```

```python
df.head()
```

```python
bool_series = df["total_bill"] > 40
```

```python
df[bool_series]
```

```python
df[df["size"] > 5]
```

```python
# and -> & когда оба условия выполняются
# or -> | когда хотя бы одно из условий выполняется
```

```python
df[(df["total_bill"] > 40) & (df["size"] > 5)]
```

```python
df[(df["day"] == "Sun") | (df["day"] == "Sat")]
```

```python
options = ["Sat", "Sun"]
```

```python
df[df["day"].isin(options)]    #  df[df["day"].isin(["Sat", "Sun"])]
```
