**Стилизация в Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0, 1,1])

ax.plot(x, x, label="X vs X")

ax.plot(x, x**2, label="X vs X**2")

ax.legend(loc=(1.1, 0.5))
```
