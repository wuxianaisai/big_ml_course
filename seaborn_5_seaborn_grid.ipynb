```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Data/StudentsPerformance.csv")
```

```python
df.head()
```

```python
sns.catplot(data=df, x="gender", y="math score", kind="box", hue="gender", col="lunch", row="test preparation course")
```

```python
pg = sns.PairGrid(data=df, hue="gender")
pg = pg.map_upper(sns.scatterplot)
pg = pg.map_diag(sns.histplot)
pg = pg.map_lower(sns.kdeplot)
pg = pg.add_legend()
```
