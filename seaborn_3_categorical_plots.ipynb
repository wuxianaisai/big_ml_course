**Distribution Plots Ð² Seaborn**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Data/dm_office_sales.csv")
```

```python
df.head()
```

```python
df["division"].value_counts()
```

```python
plt.figure(figsize=(9,4), dpi=200)
sns.countplot(data=df, x="division", hue="division")
```

```python
df["level of education"].value_counts()
```

```python
plt.figure(figsize=(9,4), dpi=200)
sns.countplot(data=df, x="level of education", hue="division", palette="Set2")
```

```python
plt.figure(figsize=(9,4), dpi=200)
sns.barplot(data=df, x="level of education", y="salary", estimator=np.mean, errorbar="sd", hue="division")
plt.legend(loc=(1.01, 0.2))
```

```python
df = pd.read_csv("/content/drive/MyDrive/Data/StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(9,4), dpi=200)
sns.boxplot(data=df, y="math score", x="test preparation course", hue="test preparation course")
```

```python
plt.figure(figsize=(9,4), dpi=200)
sns.boxplot(data=df, y="reading score", x="test preparation course", hue="test preparation course")
```

```python
plt.figure(figsize=(9,4), dpi=200)
sns.boxplot(data=df, y="reading score",
            x="parental level of education",
            hue="test preparation course", palette="Set2")
plt.legend(loc=(1.02, 0.4))
```

```python
plt.figure(figsize=(9,5), dpi=200)
sns.violinplot(data=df, x="reading score", y="parental level of education", hue="parental level of education")
```

```python
plt.figure(figsize=(9,6), dpi=200)
sns.violinplot(data=df, x="reading score", y="parental level of education",
               split=True, inner="quartile")
```

```python
plt.figure(figsize=(9,6), dpi=200)
sns.violinplot(data=df, x="reading score", y="parental level of education",
               bw=0.1)
```

```python
plt.figure(figsize=(9,6), dpi=200)
sns.swarmplot(data=df, x="math score", y="gender", size=2, hue="test preparation course")
plt.legend(loc=(1.01, 0.45))
```

```python
sns.boxenplot(data=df, x="math score", y="test preparation course", hue="gender")
```
